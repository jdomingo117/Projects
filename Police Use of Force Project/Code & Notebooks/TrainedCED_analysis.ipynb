```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file = 'clean_data_v1.csv'
data = pd.read_csv(file, index_col = 0)
data.head()
```

```python
for i in data.columns:
    print(i)
```

```python
ced = data[['TrainedCED']]
ced
```

```python
ced = data[['TrainedCED']]
ceddict = dict(ced['TrainedCED'].value_counts())
yes = ceddict.get(1)
no = ceddict.get(0)
ced = pd.DataFrame([yes, no], index = ['Trained', 'Not Trained'], columns = ['Counts'])
ced
```

```python
# Visualize data
plt.figure(figsize=(8, 6))
ax = sns.set_style('darkgrid')
ax = sns.barplot(x=ced.index, y = ced['Counts'], palette = 'rocket')
ax.set_xlabel('Number of Officers', fontsize = 15)
ax.set_ylabel('Counts', fontsize = 15)
ax.axes.set_title('Number of Officers trained in CED (Taser)', fontsize = 20)
plt.xticks()
plt.tick_params(labelsize = 13)
plt.tight_layout()
```

```python
pie, ax = plt.subplots(figsize = (6,6))
labels = ced.index
colors = ['#ff9999','#66b3ff']
plt.pie(x = ced['Counts'], autopct="%.1f%%", labels = labels,
       pctdistance = 0.8, colors = colors, labeldistance = 1.05, textprops = {'fontsize': 13}, radius = 0.5)
plt.title('Proportion of Officers trained in CED (in 149,895 cases)', fontsize = 15)
ax.axis('square')
plt.tight_layout()
```

# Conclusions

- In 149,895 cases, only 56,450 officers were trained in using a taser.
- 91,445 officers were not trained.
- Thats only 38.2% of officers trained in taser use.
