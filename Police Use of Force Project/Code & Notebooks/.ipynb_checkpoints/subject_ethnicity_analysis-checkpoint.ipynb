```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file = 'clean_data_v1.csv'
data = pd.read_csv(file, index_col = 0)
data.head()
```

```python
for i in data.columns:
    print(i)
```

```python
ethnic = data[['SubjectEthnicity: Asian',
                'SubjectEthnicity: Black',
                'SubjectEthnicity: Chinese',
                'SubjectEthnicity: Unknown',
                'SubjectEthnicity: Mixed',
                'SubjectEthnicity: Other',
                'SubjectEthnicity: White',
              'Was Force Needed?']]
ethnic
```

```python
for i in ethnic.columns:
    print(ethnic[i].value_counts())
```

```python
counts = {'Asian': 19607,
         'Black':53853,
         'Chinese': 732,
         'Unknown':1666,
         'Mixed': 4741,
         'Other': 6359,
         'White': 60937}
counts = pd.DataFrame.from_dict(counts, orient = 'Index', columns = ['Counts'])
counts = counts.sort_values(by = 'Counts', ascending = False)
counts
```

```python
plt.figure(figsize = (10, 8))
ax = sns.set_style('darkgrid')
ax =sns.barplot(x=counts.index, y = counts['Counts'], palette = 'rocket')
ax.set_xlabel('Ethnicity', fontsize = 15)
ax.set_ylabel('Counts', fontsize = 15)
ax.axes.set_title('Distribution of Ethnicity in All Records', fontsize = 20)
plt.xticks()
plt.tick_params(labelsize = 13)
plt.tight_layout()
```

```python
# Pie chart
pie, ax = plt.subplots(figsize = (10, 10))
labels = counts.index
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99', '#e3a1f4', '#a1f4ea', '#edf991']
plt.pie(x = counts['Counts'], autopct="%.1f%%", explode=(0, 0, 0, 0, 0, 0, .05), labels=labels, pctdistance=0.8, colors = colors, labeldistance = 1.05, textprops = {'fontsize': 14}, radius = 0.5)
plt.title("Distribution of Ethnicity in All Records", fontsize=18)
ax.axis('square')
plt.tight_layout()
```

# Distribution of Ethnicity within all Force used records

```python
ethnic_force = ethnic[ethnic['Was Force Needed?'] == 1]
ethnic_force
```

```python
for i in ethnic_force.columns:
    print(ethnic_force[i].value_counts())
```

```python
counts_force = {'Asian': 7052,
         'Black':25441,
         'Chinese': 273,
         'Unknown':961,
         'Mixed': 2157,
         'Other': 2378,
         'White': 28578}
counts_force = pd.DataFrame.from_dict(counts_force, orient = 'Index', columns = ['Counts'])
counts_force = counts_force.sort_values(by = 'Counts', ascending = False)
counts_force
```

```python
plt.figure(figsize = (10, 8))
ax = sns.set_style('darkgrid')
ax =sns.barplot(x=counts_force.index, y = counts_force['Counts'], palette = 'rocket')
ax.set_xlabel('Ethnicity', fontsize = 15)
ax.set_ylabel('Counts', fontsize = 15)
ax.axes.set_title('Distribution of Ethnicity in Incidents which Required Force', fontsize = 20)
plt.xticks()
plt.tick_params(labelsize = 13)
plt.tight_layout()
```

```python
# Pie chart
pie, ax = plt.subplots(figsize = (10, 10))
labels = counts_force.index
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99', '#e3a1f4', '#a1f4ea', '#edf991']
plt.pie(x = counts_force['Counts'], autopct="%.1f%%", explode=(0, 0, 0, 0, 0, 0, .05), labels=labels, pctdistance=0.8, colors = colors, labeldistance = 1.05, textprops = {'fontsize': 14}, radius = 0.5)
plt.title("Distribution of Ethnicity in Incidents which Required Force", fontsize=18)
ax.axis('square')
plt.tight_layout()
```

```python
counts_combined = counts
for i in counts_force['Counts']:
    counts_combined = counts_combined.append({'Counts': i}, ignore_index = True)
counts_combined.drop(columns = ['With Force'], inplace = True)
labels = ['All Records',
          'All Records',
          'All Records',
          'All Records',
          'All Records',
          'All Records',
          'All Records',
         'Incidents where Force Required',
         'Incidents where Force Required',
         'Incidents where Force Required',
         'Incidents where Force Required',
         'Incidents where Force Required',
         'Incidents where Force Required',
         'Incidents where Force Required']
ethnicities = ['White', 'Black', 'Asian', 'Other', 'Mixed', 'Unknown', 'Chinese',
              'White', 'Black', 'Asian', 'Other', 'Mixed', 'Unknown', 'Chinese']

counts_combined['record type'] = labels
counts_combined['ethnicity'] = ethnicities
counts_combined
```

```python
plt.figure(figsize = (10, 8))
ax =sns.barplot(x='ethnicity', y = 'Counts', data = counts_combined, hue = 'record type', palette = 'rocket')
ax.set_xlabel('Ethnicity', fontsize = 15)
ax.set_ylabel('Counts', fontsize = 15)
ax.axes.set_title('Ethnic Distribution Proportions', fontsize = 20)
plt.xticks()
plt.tick_params(labelsize = 13)
plt.tight_layout()
```

# Summary DataFrame

```python
summary = counts
summary['Counts Requiring Force'] = counts_force['Counts']
summary.rename(columns = {'Counts': 'Counts out of all Records'}, inplace = True)
summary
```

```python
summary['Proportion of subset which required use of force (%)'] = round(summary['Counts Requiring Force']/summary['Counts out of all Records'],2)
summary
```

```python
summary.sort_values(by = 'Proportion of subset which required use of force (%)', ascending = False, inplace = True)
summary
```
