# Full Scope Analysis

```python
scoped = {'No Disability': 124501,
         'Mental Disability': 21133,
         'Physical Disability': 1651,
          'Both': 610}
scoped = pd.DataFrame.from_dict(scoped, orient = 'index', columns = ['Counts'])
scoped
```

```python
pie, ax = plt.subplots(figsize = (8,8))
labels = scoped.index
colors = ['#66b3ff','#ff9999','#99ff99','#ffcc99']
plt.pie(x = scoped['Counts'], autopct="%.1f%%", labels = labels,
       pctdistance = 0.8, colors = colors, labeldistance = 1.05, textprops = {'fontsize': 10}, radius = 0.5)
plt.title('Distribution of Disabilities in All Records', fontsize = 15)
ax.axis('square')
plt.tight_layout()
```

# Combining Both

```python
combined = {'Physical Disability population': 1651,
            'Mental Disability population': 21133,
            'Physical Disability force': 864,
            'Mental Disability force': 15311}
combined = pd.DataFrame.from_dict(combined, orient = 'index', columns = ['Counts'])
combined
```

```python
combined['Force'] = ['Total Population of disability',
                    'Total Population of disability',
                    'Force required',
                    'Force required']
combined['Disability'] = ['Physical Disability',
                         'Mental Disability',
                         'Physical Disability',
                         'Mental Disability']
combined
```

```python
plt.figure(figsize = (8, 6))
ax = sns.set_style('darkgrid')
ax =sns.barplot(x='Disability', y = 'Counts', data = combined, hue = 'Force', palette = 'rocket')
ax.set_xlabel('Disabilities', fontsize = 15)
ax.set_ylabel('Counts', fontsize = 15)
ax.axes.set_title('Distribution of Disabilities in Force Usage', fontsize = 20)
plt.xticks()
plt.tick_params(labelsize = 13)
```
